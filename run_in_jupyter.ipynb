```python

import psutil

logical_cores = psutil.cpu_count(logical=True)
physical_cores = psutil.cpu_count(logical=False)

print("Logical cores:", logical_cores)
print("Physical cores:", physical_cores)
cpu_count=logical_cores
```

```python
!for f in "/content/data/20x_JAO50_C10-01-01.wmv" "/content/data/20x_JAO50_C20-01-01.wmv" "/content/data/20x_JAO50_C20-02-01.wmv" "/content/data/20x_JAO50_C2_01-01.wmv" "/content/data/20x_JAO50_C2_02-01.wmv" "/content/data/20x_JAO50_C5-01-01.wmv" "/content/data/20x_JAO50_C5-01-02.wmv" "/content/data/20x_JAO50_C5_03-01.wmv"; do base=$(basename "$f" .wmv); out="/content/runs/$base"; mkdir -p "$out"; python -m holecolor.cli run "$f" --out "$out" --workers {cpu_count} && rm -f "/content/runs/$base.zip" && (cd "/content/runs" && zip -r "$base.zip" "$base"); done
```
